```python
import pandas as pd
```

```python
# 깃허브의 www.github.com/hyokwan -> python_lecture 클릭 dataset -> kopo_customerdata.csv
# 반복문, 조건문 사용 X
# 리스트의 함수, 숫자의 함수, 문자의 함수만 이용
# 1, 1, 2, 3, 5, 5, 5 -> 2, 3 => 2.5가 나와야 함. (하나만 제외하는 것이 아닌 다 제외해야 함)

csFilePath = "https://raw.githubusercontent.com/hyokwan/python-lecture/refs/heads/master/dataset/kopo_customerdata.csv"
customerData = pd.read_csv(csFilePath)
customerTotalList = list(customerData.TOTAL_AMOUNT)

# 최소, 최대 찾기
minValue = min(customerTotalList)
maxValue = max(customerTotalList)

# 최소, 최대 중복 값 체크
minCnt = customerTotalList.count(minValue)
maxCnt = customerTotalList.count(maxValue)

# 오름차순 정렬
customerTotalList.sort()

# 중복 값 만큼 인덱스로 제거
customerTotalList = customerTotalList[minCnt:]
customerTotalList.remove(maxValue)    # 최대값은 1개만 존재하기에 remove로 삭제

# 중복 값 지워졌는지 확인  -> 확인 결과 0, 0
# minCnt = customerTotalList.count(minValue)
# maxCnt = customerTotalList.count(maxValue)
# print(maxCnt)
# print(minCnt)

# 평균 구하기 (합/개수)
sumValue = sum(customerTotalList)
cntValue = len(customerTotalList)

avgValue = sumValue / cntValue

print(int(avgValue))
```
